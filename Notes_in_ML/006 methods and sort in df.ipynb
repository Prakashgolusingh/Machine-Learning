```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Machine learning/csv/tips.csv')
# what if we want to grab only last four digits of cc no. since it is not a
# string we need to first convert it into string then str(int)[-4:]
# Similarly we can apply on the column of the df
# we have .apply() method to call any custom python fn of our own to every row in a series.
#=======================last four digit fn=============================
def last_digit(num):
  return(int(str(num)[-4:]))
def yelp(num):
  if(num<10):
    return '$'
  elif(num>=10 & num<30):
    return '$$'
  else:
    return '$$$'
df['last digit'] = df['CC Number'].apply(last_digit)
print(df.head(2))
print(df.info())
df['yelp'] = df['CC Number'].apply(yelp)
print(df.info())
print(df.head(2))
df['last digit']=df['last digit']*2
print(df.head(2))
```

```python
# Now we have two way to use multiple column value in a fn to get a new column.
# 1. we can use lambda method.
# 2. We can use vectorization method.
def quality(total_bill,tip):
  if(tip/total_bill>0.25):
    return "Generous"
  else:
    return "other"
```

```python
# by using lambda fn
df['quality']=df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
print(df.head(2))

# using vectorization
df['quality']=np.vectorize(quality)(df['total_bill'],df['tip'])

# now performance of vector is much more than lambda
```
