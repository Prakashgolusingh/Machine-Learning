```python
import numpy as np
arr=np.arange(14,24)
print(arr,arr[8])
print(arr[1:4],arr[:3]) #we use to slice array
# broadcast 
arr[:4]=100
copy_arr=arr[:8] #this will not create a new one instead create a pointer.
cp_arr=arr.copy()
print(arr)
arr[:]=00
print(arr,copy_arr,cp_arr)
```

```python
arr=np.random.randint(1,100,16)
arr=arr.reshape(4,4) #converted an array into a matrix of 3x3
print(arr,arr.shape,arr.reshape(2,8))
print(arr[1,2],arr[3][2]) # by these two way we can access an element of arr.
# how to copy or slice a part of matrix ,and assign to a variable?
print(arr[:2,1:]) # it exclude upper limit.(:2) means from 0 to 1.
print(arr[:,:1]) #this print colmn 1.
```

```python
boolean_m=arr>29
print(boolean_m)
print(arr[boolean_m])# it create an array of element satisied the condition.
print(arr[arr>45]) #conditional selection.
```

```python
arr=np.arange(4,34,3)
print(arr)
print(arr+5,arr-3)
print(arr*arr,arr-arr)
# if we divide with zero then it gives warning with print with nan in numy
# without numpy it gives error istead of warning
print(np.sqrt(45))
print(np.sin(arr))
print(np.sum(arr))
print(arr.sum(),arr.max(),arr.min(),arr.var(),arr.std())
```

```python
arr=(np.arange(0,25).reshape(5,5))
print(arr)
print(arr.sum(axis=0),arr.sum(axis=1)) 
# axis =0 means sum of all rows columnwise other word sum of each column
# axis=1 means sum of all rows
```
