```python
import numpy as np
import pandas as pd

print(np.nan)
print(pd.NaT)
```

```python
# This comparision never gives correct result 
print(np.nan==np.nan)
```

```python
# we need to use is comparision keyword to check for missing vlaues.
print(np.nan is np.nan)
# we can assign np.nan to a variable and then could check with ease.
myvar = np.nan
print(myvar is np.nan)
```

```python
# To check how many values are nan or null in dataframe.
df=pd.read_csv('/content/drive/MyDrive/Machine learning/csv/movie_scores.csv')
print(df.head())
print(df.isnull())
# This is how we select the row with pre_movie_score as null in that rows.
print(df[df['post_movie_score'].isnull()])
```

```python
# This is how we
print(df.notnull())
print(df[df['pre_movie_score'].notnull()])
```

```python
# conditional filtering
print(df[df['pre_movie_score'].notnull() | df['first_name'].notnull()])
print(df['post_movie_score'].notnull())
```

```python
# To drop all rows which contains atleast one null values that is nan
# We use dropna() for same.
print(df.dropna())
print(df['post_movie_score'].dropna())
var=df['post_movie_score'].dropna().mean()
print(var)
```

```python
# here we use thresh value in dropna
# by default thresh have a value 1
# which means that if a row contain atleast 1 not null value then we accept that row. 
# we provide a value to thresh let say x
# then we eliminate a row containing not null value less than x.
print(df.dropna(thresh=4))
print(df.dropna(thresh=5))
```

```python
# In dropna by default axis = 0 which means row
# Now in this df each column have atleast one null value so it eliminate all colmn.
print(df.dropna(axis=1))
```

```python
# drop based on any one or more than one specific attribute containing null value
print(df.dropna(subset=['last_name']))
# we it drop all rows whose last_name attribute contains null value.
```

```python
print(df.dropna(subset=['pre_movie_score'])['pre_movie_score'].mean())
```

```python
# Now what if we want to fill something at null position
print(df.fillna("jack"))
# by this way we can fill a column null value to a arbitrary value.
df['pre_movie_score']=df['pre_movie_score'].fillna(0)
print(df)
```

```python
# by this way we can fill a column null value with it's mean value.
df['post_movie_score']=df['post_movie_score'].fillna(df['post_movie_score'].mean())
print(df)
```

```python
# This is best way to fill numeric column null value with column mean value.
# It will automatically do it for you.
print(df)
print(df.fillna(df.mean()))
```

```python
# It also work on numeric column
# It takes the avg value of just above and just below value and replace with nan.
df.interpolate()
```
