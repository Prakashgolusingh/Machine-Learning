```python
import numpy as np
import pandas as pd
myindex = ['a','b','c'] # list which latter used as index in series
mydata = [2,3,4] # list which will became data for series
# creating pandas series
myser=pd.Series(data=mydata,index=myindex)
#pd.Series(mydata,myindex)->as long as it is fomat don't need to tell data,idx
print(myser)
#print(pd.Series(data=myindex)) # by default idx start from 0 ,1,2 and so on
print(myser['a'],myser[1]) # by default idx of will same as in array along with 
# default  myindex also became index of myser and can get value by any of both idx.
#==========================================================================
# if we have a dictionary panda Series is smart enough to detect index and value
ages={'sam':34,'frank':23,'spike':12}
print(pd.Series(ages)) # this will autmatically detect.
print(myser.keys()) # this returns list of all keys in that series.
```
