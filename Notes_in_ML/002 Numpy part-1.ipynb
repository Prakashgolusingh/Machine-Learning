```python
import numpy as np

# array of vector type
mylist=[1,2,3,4]  # two type of array in numpy 1.vector(1D) 2.matrix(2D)
print(type(mylist))
print(np.array(mylist)) # Convert 1D list into 1D ndarray 
myarr=np.array(mylist)
print(type(myarr))
```

```python
# matrix array
my_matrix=[[1,2,3],[4,5,6],[7,8,9]]
print(my_matrix[2][1])
print(type(my_matrix))
mymatrix=np.array(my_matrix) # Convert a 2D List into 2D ndarray
print(type(mymatrix))
print(mymatrix)
print(mymatrix[2][1])
print(mymatrix[2,1])
```

```python
# built-in fn of numpy
# ===========================np.arange()==========================

# Print counting no from 0 to 9.
print(np.arange(0,10)) # 1D array,excluding 10

# print counting no from 0 to 100 with a gap of 10.
print(np.arange(0,101,10)) # 1D array with gap between elements is 10
```

```python
# np.zeros
arr_zero=np.zeros(5)  # it makes a ndarray of float type of all values zero
print(arr_zero)
print(type(arr_zero))
mat_zero=np.zeros((3,4)) # it creates a matrix of float with all values zero
print(mat_zero)
```

```python
# np.ones
mat_one=np.ones((3,4))# it creat matrix of float type and one as its entity
print(mat_one)

# np.linspace
print(np.linspace(0,10,4)) # 0 start, 10 end and 4 means 4 element in the array 
# with (end-start) equally divided into 4 parts

# np.eye()  ->  to create identity matrix which is a square matrix too
print(np.eye(3))
```

```python
print(np.random.rand(3)) # random.rand gives a floating no form [0 ,1] in array
print(np.random.rand(3,2))
print(np.random.randint(1,50,5)) # return 5 random no b/w 1 to 50
np.random.seed(40) # it creates a set of random ,each time same random no.
# anywhere if we use these seed with same value get the same set of random no.
print(np.random.rand(4))
```

```python
# we can reshape a matrix with new have same entity as that of previous one.
arr=np.arange(1,101,10)
print(arr)
np.random.seed(43)
arr=np.random.randint(1,100,10)
print(arr.reshape(2,5))
print(arr.shape)
print(arr.max(),arr.min()) # min() and max() returns the min max value of arr
print(arr.argmax(),arr.argmin()) # return index of min max elements
print(arr.shape)
print(arr.reshape(5,2).shape)
```
